# Steepest Descent

```python
def steepest_descent(x):

  alpha = 0.02

  x1 = x[0]
  x2 = x[1]

  # Derivative w.r.t x1
  x11 = 4*(x1 - x2)**3 +12 *x2 - 1 

  # Derivative w.r.t x2
  x22 = (-4)*(x1 - x2)**3 +12 *x1 - 1 
  g = np.array([x11, x22]) 

  x_1 = x - (alpha * g)

  return x_1
```

```python
p1 = []
p2 = []

x01 = np.array([0.55, 0.7])
x02 = np.array([-0.9, -0.7])

for i in range(10):
  p1.append(list(x01))
  x01 = steepest_descent(x01)

for i in range(10):
  p2.append(list(x02))
  x02 = steepest_descent(x02)
```

## Newtons Method

```python
def Newtons_Method(q, v):
    a = []
    x1 = q
    x2 = v
    for i in range(10):
              X_0 = [x1, x2]
              a.append(X_0)

              # Derivative w.r.t x1
              x11 = 4*(x1 - x2)**3 +12 *x2 - 1
              
              # Derivative w.r.t x2
              x22 = (-4)*(x1 - x2)**3 +12 *x1 - 1
              g = [x11, x22]
                  
              x111 = 12*(x1 - x2)**2
              x12 = (-12)*(x1-x2)**2 + 12
                     
              # Hessian Matrix
              F = [[x111, x12], [x12, x111]] 
                  
              #Inverse of Hessian
              F1 = np.linalg.inv(F)

              x_1 = X_0 - np.dot(F1, g)

              x1 = x_1[0]
              x2 = x_1[1]
    return a

      
nm_1 = Newtons_Method(x01[0],x01[0])
nm_2 = Newtons_Method(x02[0],x02[1])
```
